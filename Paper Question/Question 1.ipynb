```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
X_train = np.array([[1], [2], [3], [4], [5], [6], [8], [9]])
y_train = np.array([4000, 5000, 6000, 7000, 8000, 10000, 18000, 25000])
print(y_train)
```

```python
degree = 3 # Degree of the polynomial
poly_features = PolynomialFeatures(degree=degree)
X_train_poly = poly_features.fit_transform(X_train)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train_poly, y_train)
```

```python
X_test = np.array([[7], [9], [10]])
print(X_test)
```

```python
X_test_poly = poly_features.transform(X_test)
```

```python
y_pred = model.predict(X_test_poly)
```

```python
plt.scatter(X_train, y_train, color='black', label='Actual Salaries')
plt.scatter(X_test, y_pred, color='red', label='Predicted Salaries')
plt.xlabel('Pay Scale')
plt.ylabel('Salary')
plt.title('Actual vs Predicted Salaries')
plt.legend()
plt.show()
```

```python
mse = mean_squared_error(y_train, model.predict(X_train_poly))
rmse = np.sqrt(mse)
r2 = r2_score(y_train, model.predict(X_train_poly))
```

```python
# Print predicted salaries
for i in range(len(X_test)):
 print(f"Salary for pay scale {X_test[i][0]}: {y_pred[i]}")
```

```python
print("R-squared Score:", r2)
```
