```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

x = np.array([2,3,4,5,6,7])
y = np.array([60,70,80,85,90,95])
print(f"Hourse Study:{x} \n Final Exam score :{y}")
```

```python
XMean=np.mean(x)
yMean=np.mean(y)
print(f"XMean :{XMean}\n yMean:{yMean}")
```

```python
covXY=0
varX=0
n=len(x)
for i in range(n):
    covXY+=(x[i]-XMean)*(y[i]-yMean)
    varX+=(x[i]-XMean)**2
m=covXY/varX
print(covXY)
print(varX)
print(f"slope is :{m}")
```

```python
c=yMean-(m*XMean)
print(f" value of c is :{c}")
```

```python
xPredict=m*x+c
print(xPredict)
```

```python
plt.scatter(x,y,color="red",label="scatter")
plt.xlabel("Hours Studied")
plt.ylabel("Final exam score")
plt.scatter(x,xPredict, color="blue",label="line")
plt.legend()
plt.show()
```

```python
plt.plot(x,y,label="scatter")
plt.plot(x,xPredict,label="Line")
plt.legend()
plt.show()
```

```python
rNum=0
rDen=0
for i in range(n):
    rNum+=(xPredict[i]-yMean)**2
    rDen+=(y[i]-yMean)**2
r=rNum/rDen
print(r)
print(f"Accuracy is: {round(r*100)}%")
print(f"Error is: {round(100-r*100)}%")
```
