```python
import numpy as np
import pandas as pd
import  matplotlib.pyplot as plt
x=np.array([1,2,3,4,5,6,8,9])
y=np.array([10,25,32,43,50,80,78,82])
print(f"Hours Studied: {x}{x}\n\nFinal Exam Score: {y}{y}")
```

```python
xMean=np.mean(x)
yMean=np.mean(y)
print(f"XMean :{xMean} \n yMean:{yMean}")
```

```python
covXY=0
varX=0
n=len(x)
for i in range(n):
    covXY+=(x[i]-xMean)*(y[i]-yMean)
    varX+=(x[i]-xMean)**2
m=covXY/varX
print(covXY)
print(f" slop is {m}")
```

```python
c=yMean-(m-xMean)
print(f"c:{c}")
```

```python
yProduct=m*x+c
print(yProduct)
y20=m*7+c
print(f"Price of 20 Marlera : {y20}")
y9=m*9+c
print(y9)
```

```python
plt.scatter(x,y, color = "#e91e63", label = "Scatter")
plt.xlabel("Area in Marlas")
plt.ylabel("Price per Marlas")
plt.scatter(x,yProduct, color = "#00b8d4", label = "line")
plt.legend()
plt.show()
```

```python
rNum=0
rDen=0
for i in range(n):
    rNum+=(yProduct[i]-yMean)**2
    rDen+=(y[i]-yMean)**2
r = rNum / rDen
print(f"R Square: {r}{r}")
print(f"Accuracy is: {round(r*100)}%")
print(f"Error: {round(100-r*100)}%")
```
