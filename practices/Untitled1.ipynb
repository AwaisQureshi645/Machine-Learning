```python
import pandas as pd

df=pd.read_csv('Insurance.csv', delimiter='\t')
print(df)
```

```python
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])
plt.plot(xpoints,ypoints, 'o:b', ms = 20 , mec='b', mfc='g', ls='-', linewidth='20' )
plt.show()
```

```python

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1, 'o:r')
plt.plot(y2,'*:b')
plt.xlabel("Average Sum", loc='right')
plt.ylabel("sum of the Number", loc='top')
plt.title("Fusion is calcaulate " , loc='right' )
plt.grid()
```
